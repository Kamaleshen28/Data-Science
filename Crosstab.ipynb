```python
import pandas as pd
df = pd.read_excel('https://github.com/dayanandv/Data-Science/raw/main/dataset/survey.xls')
df
```

# **Cross Tab**

```python
pd.crosstab(df.Nationality, df.Handedness)
```

```python
pd.crosstab(df.Sex, df.Handedness)
```

**With Margins**

```python
k = pd.crosstab(df.Sex, df.Handedness, margins = True)
k
```

Multi Index Column and Rows

```python
pd.crosstab(df.Sex, [df.Handedness, df.Nationality], margins = True )
```

# **Normalize**

```python
pd.crosstab(df.Sex, df.Handedness, normalize = 'index')
```

# **Aggregate function**

```python
import numpy as np
pd.crosstab(df.Sex, df.Handedness, values = df.Age, aggfunc = np.average)
```

# **Automotive Dataset eg.**

```python
headers = ["symboling", "normalized_losses", "make", "fuel_type", "aspiration","num_doors", "body_style", "drive_wheels", 
           "engine_location", "wheel_base", "length", "width", "height", "curb_weight", "engine_type", "num_cylinders", 
           "engine_size", "fuel_system", "bore", "stroke", "compression_ratio", "horsepower", "peak_rpm", "city_mpg", 
           "highway_mpg", "price"]

df2 = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data', header=None, names=headers, na_values="?" )
df2.head()
```

```python
#defining a list of models that we want to review
models = ["toyota","nissan","mazda", "honda", "mitsubishi", "subaru", "volkswagen", "volvo"]
```

```python
dfNew = df2[df2.make.isin(models)]
dfNew.head()
```

```python
#CROSSTAB btw make vs body_style

pd.crosstab(dfNew.make, dfNew.body_style)
```

```python
#Groupby

dd= dfNew.groupby(['make', 'body_style'])['body_style'].count().unstack()
dd
```

```python
# PIVOT TABLE

dfNew.pivot_table(index = 'make', columns = 'body_style', aggfunc = {'body_style' :len}, fill_value=0)
```
