```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/dayanandv/Data-Science/main/dataset/weather.csv')
df
```

# **MELT**

```python
df1 = pd.melt(df, id_vars=['day'])
df1
```

```python
df1[df1['variable'] == 'chennai']
```

```python
melted = pd.melt(df, id_vars = 'day', var_name = 'city', value_name = 'temp')
melted
```

# **UnMelt**

```python
unmelt = melted.pivot(index = 'day', columns = 'city')
unmelt
```

# **Stacking**

```python
df2 = pd.read_excel('https://github.com/dayanandv/Data-Science/raw/main/dataset/stocks.xlsx', header = [0,1])
df2
```

```python
df2.stack()
```

```python
df2.stack(level =1)
```

```python
df2_stacked.unstack()
```

# **PIVOT**

```python
df3 = pd.read_csv('https://raw.githubusercontent.com/dayanandv/Data-Science/main/dataset/weather_pivot.csv')
df3
```

```python
df3.pivot(index = 'date', columns = 'city')
```

```python
df3.pivot(index = 'date', columns = 'city', values = 'temperature')
```

# **Table**

```python
df3.pivot_table(index='city', columns='date')
```

# **Margins**

```python
import numpy as np
df3.pivot_table(index = 'city', columns = 'date', aggfunc = np.sum, margins = True)
```
